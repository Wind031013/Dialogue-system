```python
import os
import json
```

```python
def get_dir_count(path):
    try:
        count = 0
        with os.scandir(path) as entries:
            for entry in entries:
                if entry.is_file():
                    count += 1
        return count
    except FileNotFoundError:
        print(f"文件夹 {path} 不存在")
        return None
```

```python
dir_path = r"E:\论文\代码\数据处理\data"
dir_count = get_dir_count(dir_path)

def read_data(data_path):
    try:
        with open(data_path, 'r', encoding='utf-8') as f:
            lines = f.readlines()
            if len(lines) > 5:  
                cleaned_lines = lines[2:-1]
            else:
                cleaned_lines = lines
        data = ''.join(cleaned_lines)
        return data
    except FileNotFoundError:
        print(f"文件 {data_path} 不存在")
        return None
```

```python
# 插入json数据到文件
def insert_json_to_file(file_path, conversation_data, source_file):
    """将JSON数据插入到文件"""
    try:
        conversation_data['source_file'] = source_file
        with open(file_path, 'a', encoding='utf-8') as f:
            f.write(json.dumps(conversation_data, ensure_ascii=False) + ',' + '\n')
        print(f"成功保存数据到 {file_path}")
        return True
    except Exception as e:
        print(f"保存数据到 {file_path} 时出错: {e}")
        return False
```
